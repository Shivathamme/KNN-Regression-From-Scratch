Task - KNN from Scratch


Step - 1: Load the data

```python
import pandas as pd
```

```python
data = pd.read_csv("/content/diamonds (1).csv")
data.head()
```

Step - 2: Identify input and output variables

```python
x = data.drop('price',axis = 1)
y = data['price']
```

Step - 3: Split the data - Test and Train (recommended 75:25 split)

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Step - 4: Data Preprocessing on X_train (You can use sklearn for data preprocessing)¶
Categorical Data Encoding,
Numerical Data Rescaling

```python
cat_cols = ["cut","color","clarity"]
num_cols = X_train.columns.drop(cat_cols)
X_train[num_cols]
```

```python
# Categorical Data Encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder(sparse_output = False).set_output(transform = 'pandas')
encoder
```

```python
X_train_cat = encoder.fit_transform(X_train[cat_cols])
X_train_cat
```

```python
# Numerical Data Rescaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler
```

```python
X_train_num = scaler.fit_transform(X_train[num_cols])
pd.DataFrame(X_train_num)
```

```python
import numpy as np
X_train_final = np.hstack((X_train_num, X_train_cat))
pd.DataFrame(X_train_final)
```


Step - 5: Data Preprocessing on X_test

```python
X_test_cat = encoder.transform(X_test[cat_cols])
X_test_cat  = pd.DataFrame(X_test_cat)
X_test_cat
```

```python
X_test_num = scaler.transform(X_test[num_cols])
X_test_num = pd.DataFrame(X_test_num)
X_test_num
```

```python
X_test_final = np.hstack((X_test_num, X_test_cat))
X_test_final
```

Step - 6: Build the model and predict on X_test (SCRATCH IMPLEMENTATION)

Implement KNN Algorithm from scratch and do the predictions for test data. You should not use the sklearn KNN algorithm here. Write the complete code implementation from scratch for KNN algorithm

```python
import numpy as np
```

```python
def euclidean_distance(x1,x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))
```

```python
def knn_regression_predict(X_train, y_train, X_test, k = 5):
    predictions = []
    for test_point in X_test:
        distances = []
        for i in range(len(X_train)):
            dist = euclidean_distance(test_point,X_train[i])
            distances.append((dist, y_train.iloc[i]))
        distances.sort(key = lambda X:X[0])
        k_neighbors = distances[:k]
        avg_value = np.mean([value for _, value in k_neighbors])

        predictions.append(avg_value)
    return np.array(predictions)
```

```python
y_pred_scratch = knn_regression_predict(X_train_final, y_train.reset_index(drop = True),X_test_final,k=5)
```

```python
pd.DataFrame(y_pred_scratch)
```

Step - 7: Evaluate your model

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mae_scratch = mean_absolute_error(y_test,y_pred_scratch)
rmse_scratch = np.sqrt(mean_absolute_error(y_test,y_pred_scratch))
print(rmse_scratch)
print(mae_scratch)
```

Step - 8: Train a model using sklearn KNN Algorithm and compare the results with your scratch implementation

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_sklearn = KNeighborsRegressor(n_neighbors = 5)
knn_sklearn.fit(X_train_final,y_train)
```

```python
y_pred_sklearn = knn_sklearn.predict(X_test_final)
pd.DataFrame(y_pred_sklearn)
```

```python
mae_sklearn = mean_absolute_error(y_test, y_pred_sklearn)
rmse_sklearn = np.sqrt(mean_absolute_error(y_test, y_pred_sklearn))
print(rmse_sklearn)
print(mae_sklearn)
```

```python

print(f"Scratch KNN  -> MAE: {mae_scratch}, RMSE: {rmse_scratch}")
print(f"Sklearn KNN  -> MAE: {mae_sklearn}, RMSE: {rmse_sklearn}")
```
